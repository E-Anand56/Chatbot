```python
# 1. API KEY SETUP
import google.generativeai as genai

GOOGLE_API_KEY = "AIzaSyBMc1anEjYXErwaTPwYPNA-g1hscxUPOzA"   # <-- paste your API key
genai.configure(api_key=GOOGLE_API_KEY)

# 2. LOAD PDF
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("Machine Learning.pdf")
docs = loader.load()

print(f"Total pages loaded: {len(docs)}")

# 3. SPLIT TEXT INTO CHUNKS
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(
    chunk_size=1000,
    chunk_overlap=200
)

documents = text_splitter.split_documents(docs)
print(f"Total text chunks created: {len(documents)}")

# 4. CREATE EMBEDDINGS + VECTOR DB

from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_community.vectorstores import Chroma

from langchain_community.vectorstores import Chroma
from langchain_community.embeddings import HuggingFaceEmbeddings

embeddings = HuggingFaceEmbeddings(
    model_name="sentence-transformers/all-MiniLM-L6-v2"
)

db = Chroma.from_documents(
    documents,
    embedding=embeddings
)

print("Vector DB created successfully")


# 5. USER QUERY (RETRIEVAL)
query = "what is machine learning?"

retrieved_docs = db.similarity_search(query, k=1)

print("\n--- Retrieved Context ---")
print(retrieved_docs[0].page_content)


# 6. GEMINI FINAL ANSWER (RAG)

from langchain_google_genai import ChatGoogleGenerativeAI

llm = ChatGoogleGenerativeAI(
    model="models/gemini-flash-latest",
    google_api_key=GOOGLE_API_KEY,
    temperature=0
)

prompt = f"""
Answer the question using ONLY the information below.
If the answer is not present, say "I don't know".

INFORMATION:
{retrieved_docs[0].page_content}

QUESTION:
{query}
"""

response = llm.invoke(prompt)

print("\n--- Final Answer ---")
print(response.content)
```
